Znaczna większość pojawiających się korelacji wynika z chemicznego znaczenia danych pól np.: pH, a kwasowość.
Inne to np.:
- dodatnia korelacja między jakością, a ilością alkoholu
- ujemna korelacja między kwasowością lotną, a jakością - z internetu można się dowiedzieć, że ta kwasowość głównie odpowiada za smak i zapach octu w winie
- lekko dodatnia korelacja między zawartością siarczanów, a jakością

Zastanowiało mnie nagły spadek ilości chlorków, ale jak widać jest to jedynie kwestia ułożenia histogramu

```python
df.quality.value_counts()
```

W ramce przedstawione zostały wina głównie średniej jakości - ocena 5-6, nie pojawiły się żadne bardzo słabe, ani bardzo dobre - brak ocen 0-2, 9-10

```python
sns.stripplot(data = df, x = "quality", y = "alcohol", dodge = True, size = 2)
```

Im więcej alkoholu, tym lepsze wino - średnie wina (5-6) mają ilość alkoholu zgrupowaną w 9.5, a lepsze (7-8) w 11.5

```python
sns.stripplot(data = df, x = "quality", y = "sulphates", dodge = True, size = 2)
```
